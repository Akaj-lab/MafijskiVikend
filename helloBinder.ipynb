```python
""" Kako narišem graf? - 1. del """

import numpy as np # Z `np` okrajšamo zapis namesto `numpy`

# Ustvari tabelo števil
x = np.linspace(0, 2*np.pi, 100)
# print(x)

import matplotlib.pyplot as plt # knjižnica za prikaz grafov

plt.plot( x, np.sin(x) )
```

```python
""" Kako narišem graf? - 2. del """

plt.figure(figsize=(8,6)) # Graf poljubne velikosti

plt.plot(x, np.sin(x), label='sin(x)') # dodaj opis `sin(x)`
plt.plot(x, np.cos(x), '.-') # poleg poveznae črte `-` izriši tudi posamezne točke `.` => skupaj: `.-`

plt.legend(fontsize=14) # dodaj legendo

plt.title ('Naslov grafa', size=16)
plt.xlabel('opis x-osi', size=14)
plt.ylabel('opis y-osi', size=14)

plt.grid() # nariši mrežo
plt.xlim([0, 2*np.pi]) # interval izrisa v x-smeri
plt.ylim([-1.2, 1.2]) # interval izrisa v y-smeri
```
